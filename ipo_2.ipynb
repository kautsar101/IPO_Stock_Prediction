```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

# Replace with your own path to chromedriver
path = 'C:/Users/Lenovo/Downloads/chromedriver/chromedriver.exe'

# Create a Service object
service = Service(executable_path=path)

# Initialize ChromeOptions
options = webdriver.ChromeOptions()

# Create the WebDriver instance using the Service object
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()

# List to hold stock codes
stock_codes = []

# Iterate through pages 1 to 32
for page in range(1, 33):
    while True:
        url = f"https://sahamidx.com/?view=Stock.IPO&path=Stock&page={page}"
        driver.get(url)
        
        try:
            # Find all stock code elements
            stock_code_elements = driver.find_elements(By.XPATH,
                                                        "//a[starts-with(@href, '?view=Stock.Profile&path=Stock&stock_code=')]")
            
            # Extract stock codes and append to the list
            for element in stock_code_elements:
                stock_code = element.text.strip()
                stock_codes.append(stock_code)
                print(f"Code   :{stock_code}")
            print(f"Page {page} scraped successfully.")
            break

        except:
            driver.refresh()
            
            time.sleep(2.5)
            continue

# Print the stock codes
print(stock_codes)

# Quit the driver
driver.quit()
```

```python
stock_codes = [code for code in stock_codes if len(code)==4]

# Print the cleaned list
print(stock_codes)
```

```python
import pandas as pd

# Convert the list to a DataFrame
df = pd.DataFrame(stock_codes, columns=["Stock Code"])

# Save the DataFrame to a CSV file
df.to_csv('stock_codes.csv', index=False)

print("Stock codes saved to stock_codes.csv")
```
